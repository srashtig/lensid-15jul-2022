### A demo

````python
data= dict(np.load('../data/injection_pars/training/dominik_plaw2_lensed_inj_params_include_pol_phi0.npz')) 
````

````python
i= 11 # lensed event no.
j=1 # image no.
````

````python
hp,hc = get_td_waveform(approximant = "IMRPhenomPv2",mass1=data["m1"][i],mass2=data['m2'][i],\
                        inclination=data["incl"][i],delta_t=1.0/2**12,f_lower=15,f_higher = 1000, \
                        distance =data["dist"][i][j]/10,coa_phase = data['phi0'][i])
det_h1 = Detector("H1")
det_l1 = Detector("L1")
det_v1 = Detector("V1")
end_time = data["t0"][i][j]
declination = data['dec'][i]
right_ascension = data["ra"][i]
polarization = data["pol"][i]
hp.start_time += end_time
hc.start_time +=end_time

signal = det_h1.project_wave(hp,hc,right_ascension,declination,polarization)
t_gps=data["t0"][i][j]
````

````python
print(data["m1"][i], data['m2'][i],data["incl"][i], data["dist"][i][j],data['phi0'][i])
````

````python
print(end_time, declination, right_ascension, polarization)
````

````python
pylab.plot(signal.sample_times,signal.data)
````

````python
psd_H, psd_L, psd_V = inj_psds_HLV(psd_mode='analytical')

noise_signal = inject_noise_signal(signal, psd_H,whitened=False,seed=10)
````

````python
pylab.plot(noise_signal.sample_times,noise_signal.data)
pylab.plot(signal.sample_times,signal.data)
pylab.show()
````

````python
q_msmall = (4,10)
q_mlarge = (3,7)
````

````python
max_power=plot_qt_from_ts(noise_signal, t_gps, q_mlarge , outfname='',save=False)
````

### breakdown of scripts

**create_lensed_inj_xmls.py: creates inj.xml in odir**


````python
!python ../scripts/data_generating/bayestar_skymaps/create_lensed_inj_xmls.py --help
````

````python
data= dict(np.load('../data/injection_pars/training/dominik_plaw2_lensed_inj_params_include_pol_phi0.npz')) 
index=10
img=0
n_max=data['m1'].shape[0]
outdir ='.'
odir = outdir+'/'+str(index) + '/'+ str(img)+'/'
if not os.path.exists(odir):
    os.makedirs(odir)


m1=data['m1'][index]
m2=data['m2'][index]
dL=data['dist'][index][img]
t0=data['t0'][index][img]
phi0=data['phi0'][index]
pol=data['pol'][index]
iota=data['incl'][index]
ra=data['ra'][index]
dec=data['dec'][index]

os.system('lalapps_inspinj -o %sinj.xml --m-distr fixMasses --fixed-mass1 %s --fixed-mass2 %s--t-distr uniform --time-step 7200 --gps-start-time %f --gps-end-time %f --d-distr volume --polarization %f --min-distance %f --max-distance %f --l-distr fixed --latitude %f --longitude %f --i-distr fixed --fixed-inc %f --f-lower 30 --disable-spin --coa-phase-distr fixed --fixed-coa-phase %f --waveform SEOBNRv4' %(odir,m1,m2,t0,t0+200,pol*180/np.pi,dL*1e3,dL*1e3,dec*180/np.pi,ra*180/np.pi-180,iota*180/np.pi,phi0*180/np.pi))
````

**lensed_localise.sh/unlensed_localise.sh : localise with bayestar, generates psd.xml, skymap.fits and skymap.png**

    ```
    ./create_lensed_inj_xmls.py --index ${index} --img $img --odir ${out}
    
    bayestar-sample-model-psd \
    -o ${out}'/'${index}'/'${img}'/'psd.xml \
    --H1=aLIGOZeroDetHighPower \
    --L1=aLIGOZeroDetHighPower \
    --V1=AdvVirgo


    bayestar-realize-coincs \
    -o ${out}'/'${index}'/'${img}'/'coinc.xml \
    ${out}'/'${index}'/'${img}'/'inj.xml --reference-psd ${out}'/'${index}'/'${img}'/'psd.xml \
    --detector H1 L1 V1 \
    --measurement-error gaussian-noise \
    --net-snr-threshold 2.0 \
    --min-triggers 1 \
    --snr-threshold 1 \
    -P

    bayestar-localize-coincs ${out}'/'${index}'/'${img}'/'coinc.xml -o ${out}'/'${index}'/'${img}'/'

    ligo-skymap-plot ${out}'/'${index}'/'${img}'/'0.fits -o ${out}'/'${index}'_'${img}'_'skymap.png --annotate --contour 50 90
    ./sky_injs_cart.py  -index ${index} -lensed y -odir ${odir} -indir ${odir}
