```python
tag='_git'
ML_O3a_inj_df = pd.read_csv(base_dir+'data/dataframes/O3a_psd/ML_combined'+tag+'.csv',index_col=0)
```

```python
ML_predictions_O3a_real = pd.read_csv(base_dir+'data/dataframes/O3a_events/ML_combined.csv',index_col=0)
```

```python
ML_predictions_O3a_real_pe_sky = pd.read_csv(base_dir+'data/dataframes/O3a_events/ML_combined_pe_sky.csv',index_col=0)
```

```python
bins=np.linspace(-5,0,30)

ml_stat= 'densnet_xgbsky_bayestar_mul_0'
ML_predictions_O3a_real[ml_stat+'_fpp']=get_fars(ML_predictions_O3a_real,ml_stat,ML_O3a_inj_df,ml_stat[:-2])
plt.hist(np.log10(ML_predictions_O3a_real[ml_stat+'_fpp']),bins=bins,label='ML combined', histtype='step',density=True,lw=2,ls='dashed')

ml_stat= 'xgb_pred_bayestar_skymaps'
ML_predictions_O3a_real[ml_stat+'_fpp']=get_fars(ML_predictions_O3a_real,ml_stat,ML_O3a_inj_df,ml_stat)
plt.hist(np.log10(ML_predictions_O3a_real[ml_stat+'_fpp']),bins=bins,label='ML sky ', histtype='step',density=True,lw=2,ls='dashed')

ml_stat= 'xgb_dense_QTS_0'
ML_predictions_O3a_real[ml_stat+'_fpp']=get_fars(ML_predictions_O3a_real,ml_stat,ML_O3a_inj_df,ml_stat[:-2])
plt.hist(np.log10(ML_predictions_O3a_real[ml_stat+'_fpp']),bins=bins,label='ML qts ', histtype='step',density=True,lw=2,ls='dashed')

plt.legend()
plt.xlabel('FPP(log10)')
plt.ylabel('PDF')
plt.grid()
plt.suptitle('FPPs distribution O3a real events with ML')
plt.show()
```

```python
plt.figure(figsize=(10,7))
bins=np.linspace(-5,0,30)

ml_stat= 'densnet_xgbsky_pe_mul_0'
ML_predictions_O3a_real_pe_sky[ml_stat+'_fpp']=get_fars(ML_predictions_O3a_real_pe_sky,ml_stat,ML_O3a_inj_df,'densnet_xgbsky_bayestar_mul')
plt.hist(np.log10(ML_predictions_O3a_real_pe_sky[ml_stat+'_fpp']),bins=bins,label='ML combined', histtype='step',density=True,lw=2,ls='dashed')

ml_stat= 'xgb_pred_pe_skymaps'
ML_predictions_O3a_real_pe_sky[ml_stat+'_fpp']=get_fars(ML_predictions_O3a_real_pe_sky,ml_stat,ML_O3a_inj_df,'xgb_pred_bayestar_skymaps')
plt.hist(np.log10(ML_predictions_O3a_real_pe_sky[ml_stat+'_fpp']),bins=bins,label='ML sky ', histtype='step',density=True,lw=2,ls='dashed')

ml_stat= 'xgb_pred_bayestar_skymaps'
ML_predictions_O3a_real[ml_stat+'_fpp']=get_fars(ML_predictions_O3a_real,ml_stat,ML_O3a_inj_df,ml_stat)
plt.hist(np.log10(ML_predictions_O3a_real[ml_stat+'_fpp']),bins=bins,label='ML sky (bayestar)', histtype='step',density=True,lw=2,ls='dashed')

ml_stat= 'xgb_dense_QTS_0'
ML_predictions_O3a_real_pe_sky[ml_stat+'_fpp']=get_fars(ML_predictions_O3a_real_pe_sky,ml_stat,ML_O3a_inj_df,ml_stat[:-2])
plt.hist(np.log10(ML_predictions_O3a_real_pe_sky[ml_stat+'_fpp']),bins=bins,label='ML qts ', histtype='step',density=True,lw=2,ls='dashed')


plt.legend()
plt.xlabel('FPP(log10)')
plt.ylabel('PDF')
plt.grid()
plt.suptitle('FPPs distribution O3a real events with ML using PE skymaps')
plt.show()
```
